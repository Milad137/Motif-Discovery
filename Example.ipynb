```python
import numpy as np
from MotifDetection import MotifDetect
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot
```

```python
init_notebook_mode(connected=True)
df=np.loadtxt("s.txt")
```

```python
ML=10 #Length of teh Motif
X=2.7 #Major Factor of Clustor Radius
K=1.5 #Minor Factor of Clustor Radius
MD,BSFB=MotifDetect(df,ML,K,X)
print(BSFB)
```

```python
x=np.linspace(0.0, 1.0, num=ML)
for i in MD:
    data=[]
    for j in i:
        trace=go.Scatter(x=x, y=j/j.max())
        data.append(trace)
    iplot(data, filename='line-mode')
```
